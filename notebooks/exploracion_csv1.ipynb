```python
# abrimos archivo entero para analizar el porcentaje de nulos en todas las columnas


# Supongamos que tienes un DataFrame llamado df con 214 columnas
# Para ver los porcentajes de nulos en cada columna sin truncar la visualización:
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}%'.format)  # Formato de porcentaje

# Calcular los porcentajes de nulos en cada columna
porcentaje_nulos = df_csv1.isnull().mean() * 100

print(porcentaje_nulos)
```
