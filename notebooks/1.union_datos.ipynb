### UNIFICAR LAS COLUMNAS QUE SE DIVIDEN EN VARIAS PARTES

```python
def unir_col(*args):
   """
   FunciÃ³n que nos permite crear un string con las entradas con el 
   mismo Ã­ndice, de varias columnas de una dataframe.
      
   Args: 
      Las columnas de un dataframe.
         
   Return:
      Un string 
      
   """
   lista = [str(x) for x in args if str(x) != 'nan']
   return ' '.join(lista)
```
